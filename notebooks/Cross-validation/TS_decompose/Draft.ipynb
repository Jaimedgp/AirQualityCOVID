```r
# Working directory
setwd("~/Repositories/AirQualityCOVID")

load("data/data_AQ.rda")
```

```r
aq <- data_AQ[data_AQ$site == "es0118a", ] %>%
            filter(variable == "no2") %>%
            timeAverage(avg.time = "day") %>%
            drop_na()
aq$date <- as_date(aq$date)
```

```r
aq.ts <- ts(aq$value, frequency = 365)
plot(aq.ts)
```

```r
dcmps <- decompose(aq.ts)

plot(dcmps)
```
