```r
## ERA5-Land: downloaded from https://cds.climate.copernicus.eu/cdsapp#!/dataset/reanalysis-era5-land?tab=overview
## Este dataset ofrece valores horarios de un buen número de variables meteorológicas sobre tierra, a una resolución regular de 0.1º para el período 1981-2020
## De momento, yo he descargado el período 2010-2020 (si más adelante queremos hacer algo yéndonos más atrás en el tiempo, podremos actualizar fácilmente el dataset)
## Aunque me he descargado el dato horario, ya te paso el procesado a nivel DIARIO, para que tengan la misma resolución temporal que las estaciones de AEMET
## Conversaré el dato horario por si acaso nos planteásemos hacer algo a una resolución temporal más fina más adelante

## carga de datos (de la ruta donde los hayas guardado)

#########################
## relative humiditity ##
#########################
## derivada de la temperatura a 2 m. y la temperatura del punto de rocío, de acuerdo con la fórmula de Magnus: https://earthscience.stackexchange.com/questions/16570/how-to-calculate-relative-humidity-from-temperature-dew-point-and-pressure
load("data/Curation/ERA5-Land/rh_daily_2010_2020.rda")
rh.sea = lapply(list(c(12, 1, 2), 3:5, 6:8, 9:11), function(sea) {
    climatology(subsetGrid(rh, season = sea))
})
spatialPlot(makeMultiGrid(rh.sea, skip.temporal.check = T), backdrop.theme = "coastline",
           layout = c(2,2), as.table = T, set.min = 0, set.max = 100, at = seq(0, 100, 100/20), 
         names.attr = c("DJF", "MAM", "JJA", "SON"), main = "Relative Humidity (%)")
```

```r
###############################
## solar radiation downwards ##
###############################
## esta es la variable del modelo que más se asemeja a lo que mediría un piránometro en superficie
load("data/Curation/ERA5-Land/ssrd_daily_2010_2020.rda")
ssrd.sea = lapply(list(c(12, 1, 2), 3:5, 6:8, 9:11), function(sea) {
    climatology(subsetGrid(ssrd, season = sea))
})
spatialPlot(makeMultiGrid(ssrd.sea, skip.temporal.check = T), backdrop.theme = "coastline",
           layout = c(2,2), as.table = T, set.min = 0, set.max = 350, at = seq(0, 350, 350/20), 
        color.theme = "YlOrRd",
         names.attr = c("DJF", "MAM", "JJA", "SON"), main = "Solar Radiation(W/m2)")
```
