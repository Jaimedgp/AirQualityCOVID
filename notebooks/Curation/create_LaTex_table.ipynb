```python
data = pd.read_csv("./data/Curation/checked_AQ.csv")
data["values"] = 3
```

```python
data.head()
```

```python
contamianntes = pd.pivot_table(data, 
                         index = ["site", "site_name", "Municipio", "Población"],
                         columns="Pollutant", 
                         values="values",
                        ).replace(3, "x").replace(np.NAN, "-").reset_index()
contamianntes.rename(columns={"site": "Código", "site_name": "Nombre de la Estación"}, inplace = True)
contamianntes.index += 1 
```

```python
contamianntes.to_latex("doc/Anexo/estaciones_AQ.tex",
                       column_format="c"*(contamianntes.shape[1]+1),
                       label="tab:",
                       caption="",
                       index_names=False)
```
